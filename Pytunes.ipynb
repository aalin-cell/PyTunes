```python
import numpy as np
from IPython.display import Audio
from notedictionary import note_dict_fz
from durationdict import durationDict

def generate_tone(frequency, duration, sample_rate=44100):
    t = np.linspace(0, duration, int(sample_rate * duration), False)
    tone = np.sin(2 * np.pi * frequency * t)
    return tone

def playNote(note, noteDuration, BPM=120):

    quarter_note_time = 60 / BPM

    if isinstance(noteDuration, str):
        duration = quarter_note_time * durationDict.get(noteDuration)
    else: # if number
        duration = quarter_note_time * noteDuration

    if note == "REST":
        # duration = durationDict[duration_name] * (60 / BPM)
        return np.zeros(int(44100 * duration))

    frequency = note_dict_fz[note]
    # duration = durationDict[duration_name] * (60 / BPM)

    return generate_tone(frequency, duration)

def play_sequence(sequence, BPM=120):
    song = np.array([])

    for note, duration in sequence:
        tone = playNote(note, duration, BPM)
        song = np.concatenate((song, tone))

    return Audio(song, rate=44100)
```

```python
scale = [
    ("C4", "quarter"), ("D4", "quarter"), ("E4", "quarter"),
    ("F4", "quarter"), ("G4", "quarter"), ("A4", "quarter"),
    ("B4", "quarter"), ("C5", "quarter")
]

play_sequence(scale, BPM=120)
```

```python
fur_elise_test = [
    ("E5", "eighth"), ("D#5", "eighth"),
    ("E5", "eighth"), ("D#5", "eighth"),
    ("E5", "eighth"), ("B4", "eighth"),
    ("D5", "eighth"), ("C5", "eighth"),
    ("A4", "quarter")
]

play_sequence(fur_elise_test, BPM=100)
```
