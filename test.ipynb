```python
NOTE_NAMES = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
A4_INDEX = NOTE_NAMES.index("A") + (4 * 12)


def note_frequency(note_index):
    return round(440.0 * (2 ** ((note_index - A4_INDEX) / 12)), 2)


note_dict = {}

for octave in range(0, 9):
    for semitone, note_name in enumerate(NOTE_NAMES):
        note = f"{note_name}{octave}"
        note_index = semitone + (octave * 12)
        note_dict[note] = note_frequency(note_index)
```
